#### Calculating the Z score of each columns and finding the outliers

```python
from scipy import stats

# Define a threshold for Z-score (you can adjust this based on your data)
z_score_threshold = 3

# Identify outliers for each column using Z-score
outliers = pd.DataFrame()
for column in df.columns:
    z_scores = stats.zscore(df[column])
    outliers[column] = df[column][abs(z_scores) > z_score_threshold]

# Display the rows containing outliers
print("Rows with outliers:")
print(outliers)
```

```python
# Identify outliers for each column using IQR
outliers_iqr = pd.DataFrame()
for column in df.columns:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    outliers_iqr[column] = df[column][(df[column] < (Q1 - 1.5 * IQR)) | (df[column] > (Q3 + 1.5 * IQR))]

# Display the rows containing outliers using IQR
print("Rows with outliers (IQR method):")
print(outliers_iqr)
```
